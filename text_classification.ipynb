# Project 2: Topic Classification

In this project, we'll work with text data from newsgroup postings on a variety of topics. We'll train classifiers to distinguish between the topics based on the text of the posts. Whereas with digit classification, the input is relatively dense: a 28x28 matrix of pixels, many of which are non-zero, here we'll represent each document with a "bag-of-words" model. This makes the feature representation quite sparse -- only a few words of the total vocabulary are active in any given document. The bag-of-words assumption here is that the label depends only on the words; their order is not important.

```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

# General libraries.
import re
import numpy as np
import matplotlib.pyplot as plt

# SK-learn libraries for learning.
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB

# SK-learn libraries for evaluation.
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report

# SK-learn library for importing the newsgroup data.
from sklearn.datasets import fetch_20newsgroups

# SK-learn libraries for feature extraction from text.
from sklearn.feature_extraction.text import *

import nltk
```

Load the data, stripping out metadata so that we learn classifiers that only use textual features. By default, newsgroups data is split into train and test sets. We further split the test so we have a dev set. Note that we specify 4 categories to use for this project. If you remove the categories argument from the fetch function, you'll get all 20 categories.

```python
categories = ['alt.atheism', 'talk.religion.misc', 'comp.graphics', 'sci.space']
newsgroups_train = fetch_20newsgroups(subset='train',
                                      remove=('headers', 'footers', 'quotes'),
                                      categories=categories)
newsgroups_test = fetch_20newsgroups(subset='test',
                                     remove=('headers', 'footers', 'quotes'),
                                     categories=categories)

num_test = int(len(newsgroups_test.target) / 2)
test_data, test_labels = newsgroups_test.data[num_test:], newsgroups_test.target[num_test:]
dev_data, dev_labels = newsgroups_test.data[:num_test], newsgroups_test.target[:num_test]
train_data, train_labels = newsgroups_train.data, newsgroups_train.target

print('training label shape:', train_labels.shape)
print('test label shape:', test_labels.shape)
print('dev label shape:', dev_labels.shape)
print('labels names:', newsgroups_train.target_names)
```

### Part 1:

For each of the first 5 training examples, print the text of the message along with the label.

```python
def P1(num_examples=5):
### STUDENT START ###
    '''
    This function takes as argument the number of examples, and prints both the text
        and label of the first num_examples.
    '''
    
    #Define bold and regular for nice printing
    BOLD = '\033[1m'
    REGULAR = '\033[0m'
    
    #Iterate through each example
    for example in range(num_examples):
        
        #Get the index of the train label
        train_label = train_labels[example]
        
        #print the text label based on the index
        #Bold the text label
        print(BOLD + "Text label: ", newsgroups_train.target_names[train_label] + REGULAR)
        
        #print the actual training data
        print(train_data[example])
        
        #print document delimiter
        if example < num_examples - 1:
            print('-' * 60)
        
### STUDENT END ###
P1()
```

### Part 2:

CountVectorizer turns the raw training text into feature vectors. The vectorizer has a lot of options. To get familiar with some of them, we'll answer these questions:

a. The output of the transform (also of fit_transform) is a sparse matrix: http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.sparse.csr_matrix.html. What is the size of the vocabulary? What is the average number of non-zero features per example? What fraction of the entries in the matrix are non-zero? Hint: use "nnz" and "shape" attributes.

b. What are the 0th and last feature strings (in alphabetical order)? Hint: use the vectorizer's get_feature_names function.

c. Specify your own vocabulary with 4 words: ["atheism", "graphics", "space", "religion"]. Confirm the training vectors are appropriately shaped. Now what's the average number of non-zero features per example?

```python
def P2():
    '''
    This function takes the corpus of training data, and uses CountVectorize to tokenize
        and count the number of occurrences of each word in the entire dataset. Questions
        regarding the CountVectorize is also answered.
    '''
    ### STUDENT START ###
    
    #Define bold and regular for nice printing
    BOLD = '\033[1m'
    REGULAR = '\033[0m'
    
    #Initialize a CountVectorize instance
    vec = CountVectorizer()
    
    #call fit_transform tokenize and count occurrences
    #returns a sparse matrix for the data
    X = vec.fit_transform(train_data)

    #a. 
    #What is the size of the vocabulary? 
    #What is the average number of non-zero features per example? 
    #What fraction of the entries in the matrix are non-zero?

    print(BOLD + "Part A" + REGULAR)

    #Shape gives the matrix dimensions, which is the tuple (num_examples, num_features)
    #num_features is the number of unique words in the trainin set (vocabulary)
    print("The size of the vocabulary is:", X.shape[1])
    
    #The attribute nnz is the number of non-zero values in the matrix
    #Some non-zero values may be greater than 1 if there are repeated words in a document
    #Equivalent to np.sum(X.toarray() > 0)

    #Average # non-zero features per example is X.nnz/X.shape[0]
    print("The average number of non-zero features per example:", np.round(X.nnz / X.shape[0], 4))
    
    #Fraction of non-zero entries is X.nnz / total size of the matrix (X.shape[0] * X.shape[1])
    print("The fraction of entries in the matrix that are non-zero:", np.round(X.nnz/np.multiply(*X.shape), 4))
    
    #b. 
    #What are the 0th and last feature strings (in alphabetical order)? 

    print('\n' + BOLD + "Part B" + REGULAR)

    #vectorizer's get_feature_names function will return a list of features
    #    names in alphanumerical order; we get this list and index into first
    #    and last elements
    
    print('0th feature string:', vec.get_feature_names()[0])
    print('last feature string:', vec.get_feature_names()[-1])
    
    #c. 
    #Specify your own vocabulary with 4 words: ["atheism", "graphics", "space", "religion"]. 
    #Confirm the training vectors are appropriately shaped. 
    #Now what's the average number of non-zero features per example?

    print('\n' + BOLD + "Part C" + REGULAR)
    
    #First initialize and fit a new CountVectorizer with vocabulary pre-initialized
    vec_c = CountVectorizer(vocabulary=["atheism", "graphics", "space", "religion"])
    X_c = vec_c.fit_transform(train_data)
    
    #Confirm that the shape of the training vectors are appropriately shaped
    #We should get (num_examples, 4), since our vocabulary size is 4
    assert X_c.shape == (X.shape[0], 4)
    print("Size of vocabulary should be 4, as seen from shape of sparse matrix (num_examples, vocab_size): ", X_c.shape)
    
    #Average number of non-zero features per example is non_zero features / number of examples
    #We expect this to be drastically less since there are only 4 words for which non-zero features 
    #    can appear
    print("Average number of non-zero features per example:", np.round(X_c.nnz / X_c.shape[0], 4))
    
    #d. 
    #Instead of extracting unigram word features, use "analyzer" and "ngram_range" to extract bigram and trigram character features. 
    #What size vocabulary does this yield?

    print('\n' + BOLD + "Part D" + REGULAR)

    #First initialize and fit a new CountVectorizer with ngram_range
    #ngram_range defines the upper and lower limit for n-gram sizes
    #analyzer should be set to word to that entire word token are taken. This is the default
    #Note that we use char instead of char_wb. char_wb will padd n-grams at the edge of words
    #    with a space, so we end up with three character n-grams such as " aa", including the space
    
    #bi-gram words: (2,2)
    vec_d2 = CountVectorizer(analyzer='char', ngram_range=(2,2))
    X_d2 = vec_d2.fit_transform(train_data)
    
    #Size of vocabulary is the second argument to shape.
    
    print("The size of the vocabulary with bi-gram characters:", X_d2.shape[1])

    #tri-gram words: (3,3)
    vec_d3 = CountVectorizer(analyzer='char', ngram_range=(3,3))
    X_d3 = vec_d3.fit_transform(train_data)
    
    print("The size of the vocabulary with tri-gram characters:", X_d3.shape[1])

    
    #bi-gram AND tri-gram words: (2,3)
    vec_d23 = CountVectorizer(analyzer='char', ngram_range=(2,3))
    X_d23 = vec_d23.fit_transform(train_data)
    
    #Size of vocabulary is the second argument to shape.
    
    print("The size of the vocabulary with bi-gram AND tri-gram characters:", X_d23.shape[1])

    #e. 
    #Use the "min_df" argument to prune words that appear in fewer than 10 documents. 
    #What size vocabulary does this yield?

    #min_df ignore words that appear less than a specified number of times
    #This argument is not inclusive, so to get fewer than 10, we specify 10

    print('\n' + BOLD + "Part E" + REGULAR)

    vec_e = CountVectorizer(min_df=10)
    X_e = vec_e.fit_transform(train_data)
    
    #Get vocabulary size again; we expect this to be less than the original vocabulary size
    print("The size of vocabulary with words appearing 10 times or more:", X_e.shape[1])
    
    #f. 
    #Using the standard CountVectorizer:
    #    what fraction of the words in the dev data are missing from the vocabulary? 
    #Hint: build a vocabulary for both train and dev and look at the size of the difference.
    print('\n' + BOLD + "Part F" + REGULAR)

    #Build a standard CountVectorize for dev set 
    #Note vec/X for train data already build from part a
    vec_dev = CountVectorizer()
    X_dev = vec_dev.fit_transform(dev_data)
    
    #Get the vocabulary from both the train set and the dev set
    train_vocab = vec.get_feature_names()
    dev_vocab = vec_dev.get_feature_names()

    #Take the set difference to get words in the dev words that are not in the train set
    dev_vocab_not_in_train = set(dev_vocab) - set(train_vocab)
    
    #Print fraction of missing words as: length of dev vocab not in train / dev vocab size
    print("Fraction of the words in the dev data missing from the vocabulary:",
          np.round(len(dev_vocab_not_in_train) / X_dev.shape[1], 4))
    ### STUDENT END ###
    
P2()
```

### Part 3:

For baselines models, we will use the default CountVectorizer options and report the f1 score for a k nearest neighbors classifier with optimal value for k. Also, we fit a Multinomial Naive Bayes model with the optimal value for alpha, and finally, a logistic regression model with the optimal value for the regularization strength C using l2 regularization. A few questions to think about:

* Why doesn't nearest neighbors work well for this problem?
* Any ideas why logistic regression doesn't work as well as Naive Bayes?
* Logistic regression estimates a weight vector for each class, which you can access with the coef\_ attribute. Output the sum of the squared weight values for each class for each setting of the C parameter. Briefly explain the relationship between the sum and the value of C.

```python
def P3():

    ### STUDENT START ###

    '''
    This function will train kNN, Multinomial NB, and Logistic Regression classifiers
    and evaluate performance based on f1 score. 
    '''

    #First, process the training data using the standard Countvectorizer options
    #Use fit_transform on the train data in order to perform fitting and transformation
    #However, only transform on the dev data by using the same vocabulary as from train
    
    vec = CountVectorizer()
    X_train = vec.fit_transform(train_data)
    X_dev = vec.transform(dev_data)
    
    def ParamSearchDev(classifier, param, other_params = {}, return_coef = False):
        
        '''
        This function implements a search over param for classifier
        Param is a dictionary of a single parameter with multiple values to be searched
        
        Uses globally defined data:
        Trains on: X_train, train_labels
        Evaluates performance on: X_dev, dev_labels
        
        Performance evaluated using f1 score
        
        Returns the best fit clf, and the f1 score
        '''
        
        #param can take only a single parameter
        assert len(param) == 1, 'param takes a dictionary of a single key'
        
        #Get parameter name, and values to be scanned
        param_name, values = list(param.items())[0]

        #Initialize best clf and f1 scores
        best_clf = None
        best_f1_score = -np.Inf
        
        f1_scores = []
        coefs = []
        #Iterate through all values
        for value in values:
            
            #Initialize the classifier with the parameter value
            clf = classifier(**{param_name: value}, **other_params)
            
            #Fit the classifier
            clf.fit(X_train, train_labels)
            
            #Fit the training data
            #Generate f1score with dev data
            #Store classifier/f1 score if it's better than current value
            #Note that since we have multi-class classifier, using average=weighted
            #    finds average across classes weighted by true frequencies of each label
            
            prediction = clf.predict(X_dev)
            f1score = metrics.f1_score(dev_labels, prediction, average = 'weighted')
            
            #Keep the classifier and score corresponding to the best classifier
            if f1score > best_f1_score:
                best_f1_score = f1score
                best_clf = clf
            
            f1_scores.append(f1score)
            
            if return_coef:
                assert hasattr(clf, 'coef_')
                coefs.append(clf.coef_)
                
        if return_coef:
            return (f1_scores, best_clf, best_f1_score, coefs)

        return (f1_scores, best_clf, best_f1_score)

    def visualize(x, f1_scores, title, xlabel, tilt = None, step = 1, color = True, ylabel = 'F1 score'):
        '''
        This function helps to visualize bivariate relationships 
        '''
        fig = plt.figure(figsize=(10,4))
        
        if color:
            #setup the colors so that the maximum point is colored green
            #everything else is blue
            colors = ['b'] * len(x) 
            colors[np.argmax(f1_scores)] = 'g'
        else:
            colors = 'b'
        
        plt.scatter(x, f1_scores, c = colors)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.xticks(x[::step], rotation = tilt)
        plt.show()
        
    #Train kNN classifiers
    #We will scan from 1 to 31, going at intervals of 1
    #This covers the range scanned from assigned 1, which is the odds digits
    k = {'n_neighbors':np.arange(1,32,1)}
    f1_scores, clf_knn, f1score_knn = ParamSearchDev(KNeighborsClassifier, k)
    
    #Generate a plot to visualize the f1 score for different values of k
    visualize(k['n_neighbors'], f1_scores, 'kNN: k versus F1 score', 'k')
    
    #Generate results for kNN
    print("Best n_neighbors from kNN:", clf_knn.get_params()['n_neighbors'])
    print("F1 score with best n_neighbors:", f1score_knn)
    
    #Train Multinomial Naive Bayes
    #We will scan the alpha smoothing parameter from 0 to 1, and 101 values within that range
    #This parameter is the Laplace smoothing: the additional number of observations
    #    added to each class
    #Larger values than 1 tend to decrease performance further
    alphas = np.linspace(0, 1, 101, dtype=np.float32)
    
    #Alpha cannot be 0; replace with smallest possible value sklearn will take
    alphas[0] = 1.0e-10
    
    alpha = {'alpha': alphas}
    f1_scores, clf_mnb, f1score_mnb = ParamSearchDev(MultinomialNB, alpha)
    
    #visualize f1 scores
    visualize(alpha['alpha'], f1_scores, 'MultinomialNB: smoothing (alpha) versus F1 score', 
                       'alpha', 'vertical', 3)
    
    #Generate results for Multinomial NB
    print("Best alpha from Multinomial NB:", clf_mnb.get_params()['alpha'])
    print("F1 score with best alpha:", f1score_mnb)
    
    #Train Logistic Regression
    #We will scan the C parameter from 0 to 1, and 21 values within that range
    #Higher values of C does not improve performance (does not regularize enough)
    #Since C cannot equal 0, we will set the first value to 0.01 instead
    #This parameter is the 1/lambda, the regularization parameter
    #    Lambda is multiplied by the L2 penality term to reduce the sum of squared magnitude
    #Since C = 1/lambda, the larger the C, the larger the C, the less lambda, the less the regularization 
    #    of the weights of the linear predictor in the model
    weights = np.linspace(0.00, 1.0, 21, dtype=np.float32)
    weights[0] = 0.01
    C = {'C': weights}
    f1_scores, clf_lr, f1score_lr, coefs = ParamSearchDev(LogisticRegression, C, 
                                                          {'solver':'liblinear', 'multi_class':'auto'}, return_coef=True)
    
    visualize(C['C'], f1_scores, 'Logistic regression: regularization C versus F1 score', 'C', tilt='vertical')
    
    #Generate results for Logistic regression
    print("Best C from Logistic Regression:", clf_lr.get_params()['C'])
    print("F1 score with best C:", f1score_lr)
    
    #Generate num of squares of coefs for each class (by summing over axis = 1)
    ss = [np.square(c).sum(axis = 1).tolist() for c in coefs]
    
    #Print out the sum of weights
    
    for i, c in enumerate(C['C']):
        print("C parameter: ", c, "\t|Sum of squares of weights: ", np.round(ss[i], 4))
    
    
    #Plot of the sum of weights
    #Plot a line for each class
    #The transpose of ss gives a matrix with 4 rows, each with ss weights for each class
    fig = plt.figure(figsize=(10,4))
    for i,cls in enumerate(np.transpose(ss)):
        label = newsgroups_train.target_names[i]
        plt.scatter(C['C'], cls, label = label)
    
    plt.title('Logistic regression: regularization C versus sum of squares coefficients')
    plt.xlabel('C')
    plt.ylabel('Sum of Square of Weights for each class')
    plt.legend()
    plt.show()
    ### STUDENT END ###
P3()
```

## ANSWER:

**Why doesn't nearest neighbors work well for this problem?**

This is an example of the curse of dimensionality. Our feature space is very large (26879 words from the previous question). kNN assigns the label based on feature simlarity using the Euclidean distance metric. For kNN to be effective, we need effective sampling of this feature space, which consists of 2^26879 total possibilities. Since we only have 2034 samples (some of which are not even unique), the total sampled space is extremely sparse. Since we do not have effective coverage of the total sample space, a new example might be very far from all examples, and the closest k (the ones kNN uses to vote) would not actually be representative of the characteristics in the new email. From a theoretical point of view, Daume showed that for high dimensional spaces, uniform sampling of the space results in all training examples being approximately the same distance apart. This decreases the discriminative power of the distance metric to distinguish between examples of different classes.

**Any ideas why logistic regression doesn't work as well as Naive Bayes?**

Naive Bayes performs better in a setting where the number of features is much larger than the number of training examples, as is the case here. This is because NB assumes that all of the features are independent conditioned on the label, and as a result does not attempt to learn the correlations between each of the features. In other words, if K is the number of features, NB reduces modeling the joint distribution of all features into K one-dimensional problems for each class. This is a much easier problem requiring less data, which NB can effectively solve as long as the approximation is reasonable (few or no very highly correlated features). Logistic regression on the other hand attempts to learn the joint effects of all variables by modeling the logit of the probability of a class (in a one-versus-all scheme for each class) as a linear combination of all of the features. As the sample size grows, we expect Logistic Regression to start performing much better, but in a small sample scheme, there is not enough information in the training set to fully capture these feature interactions, and at the same time generalizes to all possible emails. In fact, it has been mathematically shown that generative models such as NB reach their peak performance faster (O(log(n)) than discrimative models such as LR (O(n)), thus requiring less training examples. However, this peak performance for is lower for NB than LR, so with more training examples, we can expect LR to perform better. 


**Logistic regression estimates a weight vector for each class, which you can access with the coef_ attribute. Output the sum of the squared weight values for each class for each setting of the C parameter. Briefly explain the relationship between the sum and the value of C.**

C is equal to 1/lambda, where lambda is the coefficient of the regularization penalty (either sum of absolute value of the weights for L1 or Lasso, or sum of squares of the weights for L2 or Ridge) added to the negative log likelihood function (or cost function). Since we attempt to minimize the cost function in order to find the best parameters, the effect of regularization is that large weights increase the cost and are therefore penalized. As C increases, the penalty is proportionally decreased, allowing larger weights to appear as long as larger weights decrease the negative log likelihood is more than the penalty it provides. This is precisely what we see on the graph. The sum of squares of weights increases monotonically as C is increases. Small values of C (0.01) enforces all of the weights to be closer to 0. Note that this is true for all classes since we are performing one-vs-all classification. For example, for determining the probability that the email belongs to comp.graphics, all emails in that is not comp.graphics are treated as negative examples. We are essentially performing 4 binary classifications, which is why we get 4 different sets of sum of squared weights values for each C value.

Note: The goal of regularization is to reduce overfitting. Mathematically for LR, the larger the weights, the greater the change in estimated odds for a very small change in the variable corresponding to the weight. This is a sign of overfitting, since we should expect that very close values in any one dimension to not significantly impact the odds of being in a particular class. Having a low C values enforces all weights to be close to 0, reducing the complexity of the model, which makes them more suitable for generalization.

### Part 4:

For the logistic regression model, find the 5 features with the largest weights for each label -- 20 features in total. Create a table with 20 rows and 4 columns that shows the weight for each of these features for each of the labels. Create the table again with bigram features. Any surprising features in this table?

```python
def P4():

    ### STUDENT START ###
    '''
    This function trains a Logistic Regression model using the best C found previously
    We will then find the 5 features with the largest weights for each label,
        and create a table that shows the weight for each of these features
    The same task is performed for bigram features
    '''
    
    #UNIGRAM
    
    #Process the data just like before using default settings for CountVectorizer
    #Transform the dev data, but do not fit it (use parameters trained from train_data)
    vec = CountVectorizer()
    X_train = vec.fit_transform(train_data)
    X_dev = vec.transform(dev_data)
    
    #Fit the logistic regression model using the best C found previously
    clf = LogisticRegression(C = 0.55, solver = 'liblinear', multi_class = 'auto')
    clf.fit(X_train, train_labels)
    
    def get_feature_coefs(clf, vec):
        '''
        This function takes a logistic regression clf class, and extracts the top 5
        highest coefficients for the features for each response class of the regression
        
        A vec class from CountVectorize is also required.
        
        This function will output the word corresponding to the selected vocab,
            and the coefficient of each class corresponding to the selected vocab
        '''

        #The feature weights are stored in coefficients
        #The shape of coefficients is (# classes, # features)
        #We will want to find the 5 largest values for each class
        #To do this, we can use np.argsort
        #    which will return an array containing the index in the current array 
        #    corresponding to the sorted array. Entry -1 of the output is the index 
        #    in the current array of the largest value, and -2 is the second largest, etc.
        #    np.argsort(array)[-5:] will return the indices of the 5 largest values

        #Initialize array for storing indices of features
        indices = np.zeros((1, 20), dtype=np.int)

        #Iterate over each row, which is each class label
        for i, label in enumerate(clf.coef_):

            #Generate the largest 5 indices as described above
            #This is using the largest 5 positive weights
            #We prefer this approach because positive weights seeing this word
            #    increases the odds of being in this particular class
            #Also reverse the indices so largest is at the top
            index = label.argsort()[-5:][::-1]
            
            #Below is largest 5 by absolute value, which won't be used
            #index = np.abs(label).argsort()[-5:][::-1]
            
            #Store this in the indices array
            #The location is determined by class label index
            indices[0, i*5: i*5+5] = index

        #Get the feature (vocabulary) of the selected features from vec.get_feature_names
        vocab_selected = np.array(vec.get_feature_names())[indices]

        #Get the coefficients for each class label
        #We unravel the indices into a flat array, and use that array to
        #    index along axis=1 for all 4 rows of coefficients
        coefs_selected = clf.coef_[:, indices.ravel()]

        return(vocab_selected, coefs_selected)
    
    vocab_selected, coefs_selected = get_feature_coefs(clf, vec)
    
    #Define column names for the table
    columns = newsgroups_train.target_names

    #Define row names
    rows = vocab_selected.ravel()
    
    #Transpose the data so that it is suitable for plt.table; also round to 4 digits
    data = np.round(coefs_selected.T, 4)

    #turn off the axis, and put up a title
    fig = plt.figure()
    plt.tight_layout()
    plt.axis('off')
    plt.title("Unigram Table", fontsize = 20)
    
    #Adjust the table downward for the title to fit
    plt.subplots_adjust(left=0.2, top=2)
    
    #Define colors (color blue is coefficient > 0.2)
    colors = np.where(data > 0.2, 'lightcyan', 'w')
    
    #Define col/row colors in order to group words by the class
    #    from which they originated
    colcolors = ['whitesmoke', 'gainsboro', 'silver', 'darkgrey']
    rowcolors = np.repeat(colcolors, 5)
    
    #Define column colors 
    
    #Generate the table and use proper scaling
    table = plt.table(cellText=data, colLabels=columns, rowLabels=rows, 
                      loc='center', cellColours = colors, 
                      colColours = colcolors, rowColours = rowcolors)
    table.set_fontsize(15)
    table.scale(2, 2)
    
    #Add a caption
    fig.text(0,-0.2,
             '''
             This table shows the coefficients across all classes
             for the words corresponding to the top 5 coefficients most positive
             for each class. Rows represent the words, and columns the class.
             The first 5 rows correspond to class in column 1, etc. 
             These are colored accordingly.
             Boxes colored blue are coefficients > 0.2.
             ''',
            fontsize=12)
    
    
    #BIGRAM
    
    #Process the data just like before using ngram_range 2,2
    #Transform the dev data, but do not fit it (use parameters trained from train_data)
    vec_bi = CountVectorizer(ngram_range=(2,2))
    X_train = vec_bi.fit_transform(train_data)
    X_dev = vec_bi.transform(dev_data)
    
    #Fit the logistic regression model using the best C found previously for unigrams
    #Note this may not be the optimal value for bigrams
    clf_bi = LogisticRegression(C = 0.55, solver = 'liblinear', multi_class = 'auto')
    clf_bi.fit(X_train, train_labels)
    
    #Get the selected vocab/coefs as before
    vocab_selected, coefs_selected = get_feature_coefs(clf_bi, vec_bi)

    #Define new row names
    rows = vocab_selected.ravel()
    
    #Transpose the data so that it is suitable for plt.table; also round to 4 digits
    data = np.round(coefs_selected.T, 4)

    #turn off the axis, and put up a title
    fig = plt.figure()
    plt.tight_layout()
    plt.axis('off')
    plt.title("Bigram Table", fontsize = 20)
    
    #Adjust the table downward for the title to fit
    plt.subplots_adjust(left=0.2, top=2)
    
    #Define colors (color blue is coefficient > 0.2)
    colors = np.where(data > 0.2, 'lightcyan', 'w')
    
    #Generate the table and use proper scaling
    table = plt.table(cellText=data, colLabels=columns, rowLabels=rows, 
                      loc='center', cellColours = colors, 
                      colColours = colcolors, rowColours = rowcolors)
    table.set_fontsize(15)
    table.scale(2, 2)
    
    #Add a caption
    fig.text(0,-0.3,
             '''
             This table shows the coefficients across all classes
             for the words corresponding to the top 5 most positive coefficients 
             for each class. Rows represent the words, and columns the class.
             The first 5 rows correspond to class in column 1, etc.
             Boxes colored blue are coefficients > 0.2.
             These are colored accordingly.
             Notice that "cheers kent" was among the highest 5
             for both alt.atheism and talk.religion.misc
             ''',
            fontsize=12)

    ### STUDENT END ###
P4()
```

### Part 6: 

The idea of regularization is to avoid learning very large weights (which are likely to fit the training data, but not generalize well) by adding a penalty to the total size of the learned weights. That is, logistic regression seeks the set of weights that minimizes errors in the training data AND has a small size. The default regularization, L2, computes this size as the sum of the squared weights (see P3, above). L1 regularization computes this size as the sum of the absolute values of the weights. The result is that whereas L2 regularization makes all the weights relatively small, L1 regularization drives lots of the weights to 0, effectively removing unimportant features.

Use the TfidfVectorizer -- how is this different from the CountVectorizer? Train a logistic regression model with C=100.

Make predictions on the dev data and show the top 3 documents where the ratio R is largest, where R is:

maximum predicted probability / predicted probability of the correct label

```python
def P7():

    ### STUDENT START ###

    '''
    Uses the TfidfVectorizer instead of CountVectorizer to generate bag of words representation
    '''
    
    #Instantiate and fit the TfidfVectorizer
    #fit and transform on the train, but only transform on the dev
    vec = TfidfVectorizer()
    X_train = vec.fit_transform(train_data)
    X_dev = vec.transform(dev_data)
    
    #Instantiate and fit a logistic regression
    clf = LogisticRegression(C = 100, solver = 'liblinear', multi_class = 'auto')
    clf.fit(X_train, train_labels)
    
    #Generate prediction probabilities and predictions on the development set
    prediction = clf.predict_proba(X_dev)

    #Prediction contains the probabilities for each label for each dev set example
    #The shape is (#_dev_examples, #classes)
    #To calculate the R score, we need to extract the probability of the correct label,
    #    and divide by probability of the highest predicted label
    
    #To extract prob of correct label, for each example i, 
    #    we index row i of prediction the dev_label of that example
    #For example, example 10 retains the position for predicted probability
    #    in prediction[10] corresponding to the value of the 10th dev_label
    correct_prob = np.array([prediction[i, dev_labels[i]] for i in range(prediction.shape[0])])
    
    #To extract the prob of highest label, we can use the max function
    highest_prob = prediction.max(axis = 1)

    #Calculate the R ratio with an elemenent-wise division
    R = highest_prob / correct_prob

    #Generate a histogram just to see the results
    n, bins, patches = plt.hist(R)
    plt.yscale('log')
    plt.title("Histogram of R ratio for each dev set example")
    plt.ylabel("Log scale for counts")
    plt.xlabel("R ratio")
    plt.ylim(0,1500)
    
    #Find the centers of each bin
    bin_centers = 1/2*(bins[:-1] + bins[1:])
    
    #write the counts above each bin center
    for i,c in enumerate(n):
        if c > 0:
            plt.text(bin_centers[i], c+0.05, str(int(c)))
    plt.show()
    #Return the top 3 R ratios
    #First get the index using argsort as in Part 4
    #This will return an array of where each example would be
    #    if the array were sorted
    
    #Take just the last 3 in reverse order so the highest R ratio is displayed first
    index = np.argsort(R)[-3:][::-1]
    
    #Using the index, get the R score, example, dev_labels, and prediction
    #For the prediction, we extract the examples, and then get the argmax prob to get the 
    #    actual prediction example
    predict_probs = prediction[index]
    R_top = R[index]
    dev_data_top = np.array(dev_data)[index]
    dev_label_top = dev_labels[index]
    predict_top = predict_probs.argmax(axis = 1)
    
    #For each example, print the R ratio, example, and dev/predicted labels
    for i,example in enumerate(dev_data_top):
        print('')
        print("\033[1mR ratio\033[0m:", np.round(R_top[i], 4))
        print("Predicted label:", newsgroups_test.target_names[predict_top[i]])
        print("Actual label:", newsgroups_test.target_names[dev_label_top[i]])
        print("Predicted probabilities for each class:", np.round(predict_probs[i], 4), '\n')
        print(example)
        if i < len(dev_data_top) - 1:
            print('-' * 60)
        
    ## STUDENT END ###
P7()
```

ANSWER: 

TfidfVectorizer first performs the same thing as CountVectorizer, followed by "term-frequency times inverse document-frequency" which weights the features based on rate of occurrence in the training set. The rational behind this is that highly common terms (for unigrams, such as "a", "the", etc.) that appear very often in all documents provide little explanatory power in terms of distinguishing between classes. CountVectorizer will create a sparse matrix of raw counts, meaning these very frequent terms gets higher counts and could overshadow the frequencies of rarer terms that are information for classification. The count is multipled by the idf, which log((n)/(1+df)), where n is the number of emails, and df is the number of emails containing the word. If all emails contain the word, then the scale is approximately 0, eliminating the importance of the feature, whereas if only 1 document contains the word, the scale is log(n/2) > 1, increasing the importance of that feature.

Since the R ratio describe the maximum predicted probability versus predicted probability of the correct label (for wrongly predicted examples, this is essentially how much the prefers the most probably incorrect label versus the correct one; for correct labels, this is always 1, the minimum number), a giant number for these mislabeled examples means that the model significantly preferred the incorrect label (comp.graphics). Both of these emails discuss the Book of Mormon. One way to get around this is to significantly increase the weight of "mormon" in our sparse matrix for talk.religion.misc, or significantly downweight (or even set to 0) the importance of terms such as "ftp." This can be done by directly manipulating the weights of the trained vectorizer. 

We've applied generic domain knowledge in identifying relevant words for each class here. An alternative automated procedure might be to first perform a preprocessing step where we first extract all words that appear in two or more classes below a certain frequency (in order to not account for common "stop words"). Then, we eliminate those shared words in all but one class (the class where the word appears more often), ensuring for example that "ftp" should not be a positive signal for any class except comp.graphics. 

In addition, in the very long email, we see that "Mormon" appears throughout the entire email in multiple paragraphs, whereas computer specific terms such as "ftp" only appears once within a specific context. Instead of using normalized counts as our features, we might consider normalized squared counts, or higher order polynomial features, in order to place even larger weights on words that appear multiple times throughout the entire email rather than appearing only in a specific section.

Finally, using l1 regularization with a small C should also eliminate outlier cases, since this method enforces that many of the weights for each class to be 0. We would expect that terms such as ftp that appears only a few times in talk.religion.misc emails would receive a weight of 0 under this regularization scheme.
