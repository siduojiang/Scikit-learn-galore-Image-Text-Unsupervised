### Part 2:

We produce k-Nearest-Neighbors model with k = [1,3,5,7,9], evaluate and show the performance of each model. For the 1-Nearest Neighbor model, we show precision, recall, and F1 for each label.


```python
def P2(k_values):
    '''
    This function Takes in a list of k_values, trains a kNN classifier for each.
    Evaluates overall performance for each of the k values 
    Evaluates per label performance using precision, recall, and F1 for k=1 label
    '''
    ### STUDENT START ###
   
    #We will keep a list of accuracies for plotting at the end
    accuracies = []
    
    for k in k_values:
        
        #Initiate KNeighborClassifier object, and fit to mini_train
        clf = KNeighborsClassifier(n_neighbors=k)
        clf.fit(mini_train_data, mini_train_labels)
                        
        #Append accuracy to the ongoing list accuracy
        accuracy = clf.score(dev_data, dev_labels)
        #print("Accuracy with k = %d:" %k, accuracy)
        accuracies.append(accuracy)
        
        #If k = 1, use classification_report to get precision, recall, F1; return as dictionary
        if k == 1:
            clf_report = classification_report(dev_labels, clf.predict(dev_data), output_dict=True)
            clf_print = classification_report(dev_labels, clf.predict(dev_data))
            
            
    ######## Accuracy report #########
    
    #Print each accuracy for each k
    print("Mean classification accuracy with:\n")
    for i, k in enumerate(k_values):
        print("k = %d:" %k, accuracies[i])
    
    #Generate scatter plot to show accuracies for each k
    plt.scatter(k_values, accuracies)

    #Set the x axis limits, gridlines and tick labels
    plt.xlim(0,10)
    plt.xticks(np.arange(1, 10, 2))
    
    #Print the accuracy value for each data point
    for i, label in enumerate(accuracies):
        plt.annotate(label, (k_values[i] - 0.5, accuracies[i] + 0.002))
        
    #Label the accuracy graph
    plt.title("kNN Classification Accuracy as a Function of k")
    plt.ylabel('Mean accuracy')
    plt.xlabel('k')
    plt.show()
    
    
    ########## k = 1 classification report #########
          
    #Print the default classification report
    print("Classification report for KNeighborsClassifier with k = 1")
    print(clf_print)
    
    #For k=1, store information for precision, recall, and F1 score
    k1NN_report = {
        'Precision': [],
        'Recall': [],
        'F1-Score': []
    }
    
    #for each label, extract the corresponding value from clf_report above
    labels = clf_report.keys()
    
    for label in labels:
        k1NN_report['Precision'].append(clf_report[label]['precision'])
        k1NN_report['Recall'].append(clf_report[label]['recall'])
        k1NN_report['F1-Score'].append(clf_report[label]['f1-score'])
    
    #Generate plots for metric, showing all labels on the x-axis
    fig = plt.figure(figsize = (12, 4))
    
    #Entire figure title
    fig.suptitle('Performance of 1NN classifier using 3 metrics for all digits', fontsize=15)

    #iterate for each metric, generating the scatter plot
    for i,metric in enumerate(k1NN_report.keys()):
        ax = plt.subplot(1, 3, i+1)
        ax.scatter(np.arange(0,10), k1NN_report[metric][:10])
        ax.xaxis.set_ticks(np.arange(0,10))
        ax.set_ylim(0.75, 1)
        ax.set_title(metric)
        ax.set_xlabel('digit label')
    
    fig.tight_layout()
    fig.subplots_adjust(top=0.8)
    
    ### STUDENT END ###
    
k_values = [1, 3, 5, 7, 9]
P2(k_values)
```

For k=1, the most difficult digit is '8,' which has the lowest F1-score. The F1-score is the best metric to use here, as it is defined as:

2 * (precision * recall) / (precision + recall)

This can be thought of as a weighted average of the precision and recall, and is bounded by 0 (when either or both of precision or recall is 0) and 1 (when both precision and recall are 1). 

Due to the symmetry of the calculation, the relative contribution of precision and recall to the F1 score is the same. This is desirable as we have no reason to favor precision over recall, or vice-versa. Consider the digit 8. We would favor precision (TP/(TP+FP)) when the cost of false positive is high, where a false positive is incorrectly recognizing other digits as 8. We would favor recall (TP/(TP+FN)) when the cost of a false negative is high, where false negative is incorrectly recognizing an 8 as another digit. There is no reason to favor one outcome or another, so we evaluate performance based on the F1 score.

1-Nearest Neighbor models using training data of various sizes.


```python
def P3(train_sizes, accuracies):

### STUDENT START ###

    '''
    This function takes a list of train_sizes and trains a k=1 classifier on training data of each size
    It then evaluates dev set performance on each classifier in terms of accuracy
    This function assumes that train_data, from which slices are taken, are already randomly sorted
        which is the case for this problem
    '''

    #Initialize KNN object. Each time fit is call, the previous model is overwritte.
    #Therefore we only need a single instance
    clf = KNeighborsClassifier(n_neighbors=1)

    #Store the times
    times = []
    for train_size in train_sizes:
        
        #Since the training data has been randomly shuffled,
        #we can simply take the first train_size number of values as training data
        #We will fit on the training data, then evaluate on dev data
        clf.fit(train_data[:train_size], train_labels[:train_size])
        
        #Record time it takes the measure performance
        start_time = time.time()
        predicted_values = clf.predict(dev_data)
        end_time = time.time()
        
        #calculate and score the accuracy and time elapse of the model
        train_size_accuracy = clf.score(dev_data, dev_labels)
        total_time = end_time - start_time
        accuracies.append(train_size_accuracy)
        times.append(total_time)
        print("For size %d: " %train_size, 
              "Accuracy:", train_size_accuracy,
              ", time: %0.4f seconds" %(total_time))

    #Generate a data in order to to display the output in a table
    data = list(zip(train_sizes, np.round(accuracies, 4), np.round(times, 4)))
    columns = ['Training size', 'Mean Accuracy', 'Time (seconds)']

    #turn off the axis
    plt.axis('off')
    
    #Generate the table
    table = plt.table(cellText=data,colLabels=columns,loc='center')
    table.set_fontsize(15)
    table.scale(3, 3)
    
    plt.show()
### STUDENT END ###

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25600]
accuracies = []
P3(train_sizes, accuracies)
```

### Part 4:

We now train logistic regression models that predicts accuracy of a 1-Nearest Neighbor model given training set size. We show the $R^2$ of the regression model and the accuracies it predicts for training set sizes 60000, 120000, and 1000000. We will then apply some transformation that makes the predictions more reasonable, and show $R^2$ of the improved regression model and the accuracies it predicts for training set sizes 60000, 120000, and 1000000.

```python
def P4():

### STUDENT START ###

    '''
    This function uses the accuracy trained from the previous problem (stored as a global list from P3)
    This function trains a linear regression function to predict accuracy from dataset size
    We then evaluate why this model is wrong
    Finally, we apply the logit transform in order to map a continuous prediction (from linear regression)
        onto the support from 0 to 1, which is the range of values accuracy can take.
    '''
    #LinearRegression requires reshaping our data so that it is a (n,1) ndarray
    X = np.reshape(train_sizes, (-1, 1))
    
    #First, we train a regression model on the raw features
    
    clf = LinearRegression()
    clf.fit(X, accuracies)
    
    print("\033[1mUsing raw data without transformation")
    print("\033[0mR^2 of the untransformed model: %.4f" %clf.score(X, accuracies))
    
    large_datasets = np.array([60000, 120000, 1000000]).reshape(-1,1)
    print("Predicted accuracies for 60000, 120000, and 1000000:", 
          clf.predict(large_datasets).round(3))

    #Close any previous plots
    plt.close()
    
    #Plot scatter plot of the data
    plt.scatter(X, accuracies, label = 'actual accuracy')
    
    #Plot the regression fit line
    plt.plot(X, clf.predict(X), label = 'predicted accuracy')
    plt.xlabel('Dataset Size')
    plt.ylabel('Accuracy')
    
    #Generate relevant labels
    plt.legend(loc = 4)
    plt.title("Plot of dataset size versus accuracy")
    plt.text(10000, 0.8, r'$R^2$ = %.2f' %clf.score(X, accuracies), fontsize=12)
    plt.show()
    
    #We will apply the logit transformation, which is used Logistic Regression
    #In logistic regression, the logit(probability) = linear predictor (right hand side of the equation in linear regression)
    #The transformation we apply is logit: log(accuracy / (1-accuracy))
    #This merely rescales accuracy to be always bounded between 0 and 1
    #After prediction, we transform the predicted value as 1/(1+e^-(predicted_value)) to recover accuracy
    #This inverse transformation is derived by setting logit(accuracy) = predicted_value
    #Note that predicted value is simply the value output by the linear predictor, which is related to accuracy by the logit link
   
    Y = np.array(accuracies)
    Y_transformed = np.log(Y / (1-Y))

    #In order to improve the fit, we can see from the previous plot that the shape is of a log curve as a function of x
    #In other words, Y looks like it's proportional to log(X)
    #Therefore, we will also transform X (input dataset size) using log

    X_transformed = np.log(X)
    
    #We fit and re-print the R^2
    clf.fit(X_transformed, Y_transformed)
    print("\033[1mUsing log training size with logit transformation of accuracy")
    print("\033[0mR^2 of the transformed model: %.4f" %clf.score(X_transformed, Y_transformed))
    
    #Accuracies for large dataset
    print("Predicted accuracies for 60000, 120000, and 1000000:", 
          np.round(1/(1+np.exp(-clf.predict(large_datasets))), 3))

    #Close any previous plots
    plt.close()
    
    fig = plt.figure(figsize = (16, 5))
    ax1 = plt.subplot(121)
    #Plot scatter plot of the dataset size versus accuracy
    ax1.scatter(X, accuracies,
               label = 'actual accuracy')
    
    #Plot the regression fit line
    #Prediction must be the transformed dataset sizes since it is fit
    #    to the transformed values as explained above
    ax1.plot(X, 1/(1+np.exp(-clf.predict(X_transformed))), 
             label = 'predicted accuracy',
             c = 'g')
    ax1.set_xlabel('Dataset Size')
    ax1.set_ylabel('Accuracy')
    ax1.legend(loc = 4)
    ax1.set_title("Plot of predicted accuracy and actual accuracy versus dataset size")
    
    #Generate second plot of transformed axes and show the R^2
    ax2 = plt.subplot(122)
    ax2.scatter(X_transformed, Y_transformed,
               label = 'actual accuracy')
    
    #Plot the regression fit line
    #Prediction must be the transformed dataset sizes since it is fit
    #    to the transformed values as explained above
    ax2.plot(X_transformed, clf.predict(X_transformed), 
             label = 'predicted accuracy',
             c = 'g')
    ax2.set_xlabel('Log Dataset Size')
    ax2.set_ylabel(r'$\log \frac{Accuracy}{(1-Accuracy)}$', fontsize = 15)
    ax2.legend(loc = 4)
    ax2.text(6, 3, r'$R^2$ = %.2f' %clf.score(X_transformed, Y_transformed), fontsize=12)
    ax2.set_title("Plot of logit accuracy versus log of dataset size")

### STUDENT END ###

P4()
```

Problem with using linear regression: accuracy cannot go above 1 and cannot go below 0, because it is measured as a fraction of correctly predicted examples to all examples. The predicted values for linear regression is unbounded. As a result, we must apply a transformation that bounds the response to be between 0 and 1. We chose the popular logit transform in this problem, and shows that even for large datasets the predicted accuracies does not go above 1. Also, Linear Regression assumes that the explanatory variables are linearly related to the response, which is not the case here since the relationship appears logarithmic. The significant increase of R^2 is due to the log transformation of dataset size (X). Note that linear regression was run on the transformed variables, so the R^2 is also show for the transformed variables.

### Part 6:

A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian, i.e., the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

We implement a simplified Gaussian blur by just using the 8 neighboring pixels like this: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. The blur filter will be applied in 3 ways:
- Filter the training data but not the dev data
- Filter the dev data but not the training data
- Filter both training data and dev data

```python
def P6():
    
### STUDENT START ###
    '''
    This function applies a blurring function to the images in the mini_train and dev sets
    This function then evaluates accuracy of blurring only the mini_train and dev sets, and
        both at the same time
    '''
    
    def simple_blur(image, weights=[1,1,1,1,4,1,1,1,1]):
        '''
        Takes in an image and applies a simplified blurring scheme:
            The pixel strength of the image is a weighted average of the
            value of the pixel and the 8 pixels around it
            Weight is the value fraction of the central pixel
        We must also deal with edge cases.
        For pixels in the corners, we will take a weighted average of the pixel
            and 3 pixels around it, and treat the rest of the 5 pixels as 0. 
        For pixels on the edges, we will take a weighted average of the pixel
            and the 5 pixels around it, treating the rest of the pixels valued at 0.
        We use a default weight vector where the center weight is treated 4 times as much
            as the pixels around it (corresponding to 25% of total weight: 4/(4 + 1*8))
        '''

        #reshape the image to 28 by 28 for easy reference to nearby pixels
        image = image.reshape([28,28])

        #Pad the image on all edges of both axes with 1 set of zeroes 
        #The purpose of this is to handle edge and corner cases
        #All pixels of the original image will now have 8 neighbors
        #    Edges will simply have 0 pixel values where the image doesn't exist
        #    For example, the top left corner will have 5 pixel neighbors (above and to the left)
        #        which are 0
        #New shape is 30x30, with row0, row29, col0, col29 zeroes
        image = np.pad(image, 1)
        assert image.shape == (30, 30)

        #Setup a new image
        new_image = np.zeros([28,28])

        #iterate over each i,j index, where i is row, and j is column
        #Note we index from 1 to 29 because we have padded each side with 0
        #    that doesn't belong to the image
        for i in range(1, 29):
            for j in range(1, 29):

                #Index into the 8 neighboring pixels, and then flatten the array
                pixels = image[i-1:i+2, j-1:j+2].flatten()

                #Apply linear combination of weights with the new pixel
                #Then, normalize by the sum of the weights
                new_pixel = np.dot(pixels, weights) / np.sum(weights)

                #Add modified pixel to the new image
                new_image[i-1][j-1] = new_pixel

        return new_image.flatten()

    #Apply the filter to both the mini train and the dev dataset for evaluation
    #This is done using a list comprehension over each example
    #Apply random weights seems to perform very well, and better than deterministic weights
    filtered_mini_train_data = np.array([simple_blur(train_datum) for train_datum in mini_train_data])
    filtered_dev_data = np.array([simple_blur(dev_datum) for dev_datum in dev_data])

    #Set up kNN; k = 1 for all cases, so we can reuse the same classifier
    clf = KNeighborsClassifier(n_neighbors=1)

    print("\033[1mMean accuracies for each filtering condition\033[0m")
    #Fit on non-filtered train and evaluate on non-filtered dev
    clf.fit(mini_train_data, mini_train_labels)
    print("Fit on non-filtered train, evaluated on non-filtered dev: ", clf.score(dev_data, dev_labels))

    #Fit on filtered train and evaluate on non-filtered dev
    clf.fit(filtered_mini_train_data, mini_train_labels)
    print("Fit on filtered train, evaluated on non-filtered dev: ", clf.score(dev_data, dev_labels))

    #Fit on non-filtered train and evaluate on filtered dev
    clf.fit(mini_train_data, mini_train_labels)
    print("Fit on non-filtered train, evaluated on filtered dev: ", clf.score(filtered_dev_data, dev_labels))

    #Fit on filtered train and evaluate on filtered dev
    clf.fit(filtered_mini_train_data, mini_train_labels)
    print("Fit on filtered train, evaluated on filtered dev: ", clf.score(filtered_dev_data, dev_labels))

    #Plot the non-blurred image for visualization
    ax = plt.subplot(1, 3, 1)
    ax.set_title('Original Image')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(mini_train_data[0].reshape(28,28))
    
    #Plot the first blurred image just to visualize it
    ax = plt.subplot(1, 3, 2)
    ax.set_title('Blurred Image\nCenter Weight=8')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(filtered_mini_train_data[0].reshape(28,28))
    
    #Plot "blurred" image just to visualize it
    #The purpose of this is that if the center weight is extremely large
    #    the original image will essentially be recovered
    #We see that this is indeed the case
    #This is a test case that will help diagnose any bugs in the blurring code
    ax = plt.subplot(1, 3, 3)
    ax.set_title('"Blurred" Image\nCenter Weight=1e6')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(simple_blur(mini_train_data[0], weights=[1,1,1,1,1000000,1,1,1,1]).reshape(28,28))
    
    #The blurred images seem to be correct. The location of the pixels are correct (no shifting), and the pixels
    #    in fact seem to be blurred.
    
### STUDENT END ###

P6()
```

```python
def P7():

### STUDENT START ###

    '''
    This function will build two Naive Bayes models, one with BernoulliNB (using binary features),
        and the other with MultinomialNB (using three features).
    For mapping continuous features to binary values, we will use the image mean as the threshold. 
    This is reasonable because most of the image is 0, meaning the mean will be much closer to 0 than to 1.
    This will be enough to filter out low intensity pixels, while ensuring that any pixel with signal
         is marked as 1. 
    For multinomial, we will threshold at:
        threshold 1 between white and gray: (max_pixel_intensity - min_pixel_intensity)* 1/3
        threshold 2 betweeh gray and black: (max_pixel_intensity - min_pixel_intensity) * 2/3
    '''
       
    #Generate the Bernoulli threshold set
    #This is done using np.where, where the threshold level is defined as mean of the image
    #    and set to 1 if the pixel is greater than mean, or 0 otherwise
    bernou_mini_train = np.array([np.where(image > np.mean(image), 1, 0) for image in mini_train_data])
    bernou_dev = np.array([np.where(image > np.mean(image), 1, 0) for image in dev_data])
    
    #We will fit without smoothing; smallest alpha allowed is 1.0e-10
    clf_Bernoulli = BernoulliNB(alpha = 1.0e-10)
    clf_Bernoulli.fit(bernou_mini_train, mini_train_labels)
    
    #Get our predictions on the threshold_dev set
    print("Dev set Accuracy for BernoulliNB: ", clf_Bernoulli.score(bernou_dev, dev_labels))

    
    #For multinomial, we need 2 thresholds, one between white and gray, and one between white and black
    #This is clearly a hyperparameter that can be tuned
    #For this exercise, white will first be determined by the method above:
    #    below (max_pixel_intensity - min_pixel_intensity)* 1/3
    #Black is:
    #    above (max_pixel_intensity - min_pixel_intensity)* 2/3
    #White is:    
    #    in between values
    
    def multinomial_process(images):
        
        new_images = []
        for image in images:
            #Find the range of pixel intensities
            max_minus_min = np.max(image) - np.min(image)
            
            #assign white values as 0
            #All other values are assigned as 1
            white_black_image = np.where(image > 1/3 * max_minus_min, 1, 0)
            
            #assign gray and black values; note that some values
            #that should be white will also be assigned as 1
            #We take care of this by multiplying element-wise
            #    by the previously generated white_black_image
            gray_black_image = np.where(image > 2/3 * max_minus_min, 2, 1)
            image = white_black_image * gray_black_image
            
            #Append new image
            new_images.append(image)
        return np.array(new_images)
    
    #Repeat the process for train and dev sets
    multinom_train = multinomial_process(mini_train_data)
    multinom_dev = multinomial_process(dev_data)
        
    clf_multinomial = MultinomialNB(alpha = 1.0e-10)
    clf_multinomial.fit(multinom_train, mini_train_labels)
    
    #Get our predictions on the threshold_dev set
    print("Dev set Accuracy for MultinomialNB: ", clf_multinomial.score(multinom_dev, dev_labels))

    #Print training accuracies as well
    print("\nTraining set Accuracy for BernoulliNB: ", clf_Bernoulli.score(bernou_mini_train, mini_train_labels))
    print("Training set Accuracy for MultinomialNB: ", clf_multinomial.score(multinom_train, mini_train_labels))

    #Show the original first image, binarized image, and multinom image
    
    ax = plt.subplot(1, 3, 1)
    ax.set_title('Original Image')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(mini_train_data[0].reshape(28,28))
    
    #Plot the first binarized image
    ax = plt.subplot(1, 3, 2)
    ax.set_title('Binarized Image')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(bernou_mini_train[0].reshape(28,28))
    
    #Plot the first multinomial image
    ax = plt.subplot(1, 3, 3)
    ax.set_title('Multinomial Image')
    plt.setp(ax, xticks=(), yticks=())
    ax.imshow(multinom_train[0].reshape(28,28))
### STUDENT END ###

P7()
```

ANSWER: We did not get better accuracy. NB models assume that conditional on the digit, the probability of each pixel being in a particular category (0 or 1 in Bernoulli; 0,1,2 in Multinomial) is indendependent of each other. The assumption is more strongly violated in the Multinomial case because the presence of the middle "gray" category is strongly dependent on the presence of white and black pixels around it. In other words, in the presence of a gray pixel, the neighboring pixels are more likely to be black or white rather than gray, because gray represents the edges of black pixels, or immediate values for where transitions between black and white occur.  The sharp 0,1 transitions between black and white as seen in the Bernoulli case do not violate the assumption as much. Furthermore, in the context of this problem, it doesn't make sense to create features belonging to three classes. We either have a pixel that intentionally is turned on, or a pixel that should be turned off. For example, there's no sensible "middle" category where the writer "partially" intended to draw a mark.

Another possibility is that we have yet to fine tune the hyperparameter, which is the thresholds of the model. Given that the multinomial has two possible thresholds (hyperparameters), there may very well be a set of hyperparameters where dev set performance exceeds to best single threshold in the Bernoulli case. 

Note that I've also calculated the training error to show that this not a case of overfitting. Even though we have increased model complexity, the training accuracy for MultinomialNB is still less than for BernoulliNB, which means that the multinomial has higher bias and higher variance. This simply means that the multinomial itself is less suitable for the problem than is Bernoulli. In the case of overfitting, we would expect training error to be lower.

The probability of observing a particular continuous feature given a label are modeled as Gaussians. At prediction, the probability of a label given a class is calculated using the PDF of the Gaussian evaluated at the feature value. However, the computed variances for most of features are tiny. One example of why this might be the case are edge pixels: these are pixels for which the value is close to 0 and somewhat invariant across all images because most people would rarely write in that pixel.

For such pixels, this means that when the PDF is calculated, even very small deviations from the mean value will generate large PDF values. For example, if a feature is centered at 0 (which the distribution of theta suggests that most are close to 0) with very small variance (<0.0001), an image with even a 0.1 pixel value will evaluate to a probability close to 0 in the Gaussian PDF. This means that the model is **overconfident** about the probability of this particular feature. This is problematic because very small values will quickly diminish the probability of any class, because the final P(class | features) is calculated as a product of these probabilities. This makes the model worse as evidenced by a worse training and dev set performance. Note that while we've increased the complexity of the model, the decline is not due to overfitting since the training accuracy was much lower for the GaussianNB.

To avoid this, we can artificially fix the variance (as discussed in async), or we can increase the variance of all features by adding a var_smoothing parameter. Scikit implements the latter by adding a (var_smoothing * largest variance of all features) term to every variance value, so the distribution of variance is not changed, but only shifted to the right. The distribution of theta remains the same, so this is equivalent to keeping the Gaussians of all features centered at the same location, but just stretching it out in both tails. This decreases the overconfidence of the model about the probability of observing any particular feature by spreading out the probability to the Gaussian tails (the example above would evaluate to a larger probability than 0), and which improves performance by making less extreme predictions on feature probability. Note that the means (sigma values) do not change.

### Part 10:

Because Naive Bayes produces a generative model, we can use it to generate digit images. We produce a Bernoulli Naive Bayes model and then use it to generate a 10x20 grid with 20 example images of each digit.

```python
def P10(num_examples):

### STUDENT START ###

    '''
    This function generates a Bernoulli Naive Bayes model, and uses the posterior probabilities of the features
        given a class to generate new images of digits belonging to the class. 
    '''
    
    ### STUDENT START ###

    #Generate the Bernoulli threshold set as before in part 7 (using mean pixel intensity)
    #This is done using np.where, where the threshold level is defined as mean of the image
    #    and set to 1 if the pixel is greater than mean, or 0 otherwise
    bernou_mini_train = np.array([np.where(image > np.mean(image), 1, 0) for image in mini_train_data])

    #First train the model using the best alpha parameter found previously using GridSearchCV
    NB = BernoulliNB(alpha = 0.0001)
    NB.fit(bernou_mini_train, mini_train_labels)

    #Set matplotlib schemes
    plt.rc('image', cmap = 'binary')
    fig = plt.figure(figsize = (10 * num_examples / 10, num_examples))

    #Iterate over all digits and samples
    for digit in range(10):
        for example in range(num_examples):
            
            #The generative procedure is as follows:
            #We have the probability of a pixel being 1 stored in feature_log_prob_
            #These probabilities are independent conditional on the label
            #Therefore, we can generate 784 draws from the uniform distribution
            #    and compare each draw to the probabilities. If the draw is less
            #    than the probability at that location, then the pixel is 1.
            #    This is because the probability of drawing a value from the uniform that is less than feature
            #         probability P is precisely P, the probability that the feature is 1

            #Generate 784 random numbers
            uniform_samples = np.random.rand(784)

            #Are the values from the draw less than the samples? 
            #We get this from a pairwise comparison with probabilities for each pixel
            image = (uniform_samples < np.exp(NB.feature_log_prob_[digit])).astype(np.int)

            #The index is the calculated as how many images we've produced with previous digits (digit * num_examples)
            #    plus the sample number we are on
            ax = plt.subplot(10, num_examples, (digit * num_examples) + (example + 1))
            
            # Turn off tick marks to keep things clean.
            plt.setp(ax, xticks=(), yticks=())
            ax.axis('off')
            
            ax.imshow(image.reshape(28,28))
    ### STUDENT END ###

### STUDENT END ###

P10(20)
```

The images are certainly discernible to the human eye but is much noisier than actual hand written digits. This is because the images are in a sense an ensemble average of the human hand written digits. They take all of the images and put a probability on each pixel being 1, then turns on each pixel based on its probability. Take a simple case of two pixels A,B, and two training images, where one image is all white, and one image is all black. The probability of each pixel is 0.5, and the probability of both being white or black (reproducing one of the original image) is only 2*(0.25) = 0.5. As the number of pixels increase, the probability of generating any particular image in the training set also decreases due to the increasing number of possible images. So we shouldn't be surprised they do not represent any particular training image.

### Part 11:

Recall that a strongly calibrated classifier is roughly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior probability of the predicted class is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior probability and accuracy.  

Using a Bernoulli Naive Bayes model, we evaluate performance by partition the dev set into several buckets based on the posterior probabilities of the predicted classes - think of a bin in a histogram- and then estimate the accuracy for each bucket. So, for each prediction, we find the bucket to which the maximum posterior probability belongs, and update "correct" and "total" counters accordingly. 

```python
def P11(buckets, correct, total):
    
### STUDENT START ###

    '''
    This function determines whether the Bernoulli Naive Bayes Model is calibrated.
    This function accepts a buckets argument, containing the bin edges for each bucket.
    This function also populates the number of correctly predicted examples, and the total number of examples
        belonging to each bucket.
    '''

    #Generate the Bernoulli threshold set as before in Part 7 (using mean pixel intensity)
    #This is done using np.where, where the threshold level is defined as mean of the image
    #    and set to 1 if the pixel is greater than mean, or 0 otherwise
    bernou_mini_train = np.array([np.where(image > np.mean(image), 1, 0) for image in mini_train_data])
    bernou_dev = np.array([np.where(image > np.mean(image), 1, 0) for image in dev_data])
    
    #First train the model using the best alpha smoothing parameter found previously using GridSearchCV
    #This is the optimal value for this particular binarization scheme
    NB = BernoulliNB(alpha = 0.0001)
    NB.fit(bernou_mini_train, mini_train_labels)

    #Generate the probabilities estimates of the development set
    #predict_proba will return probabilities for each label. Since we only care about
    #    the max label, we will run np.max(axis=1) to get the max label of each row
    probs = NB.predict_proba(bernou_dev)
    probs_max_label = np.max(probs, axis = 1)
    
    #Generate the prediction class of the dev set
    pred_class = NB.predict(bernou_dev)
    
    #In order to partition the dev set into buckets. np.digitize
    #    will do the binning if we supply the bins. The problem 
    #    specifies the right edges of the buckets, and np.digitize treats  
    #    the specified value as a non-inclusive bound by default. For example,
    #    since the first value is 0.5, the first bin contains [0, 0.5).
    #    Since the second value is 0.9, the second bin contains [0.5, 0.9)
    #    This is fine except for the right-most bin. We want that bin to include
    #    values that are 1. Therefore, we specify right=True. In this case, the first bin0 
    #    will include [0, 0.5], second bin1 (0.5, 0.9], last bin8 (0.9999999999999, 1]
    dev_bins = np.digitize(probs_max_label, buckets, right=True)
    
    #calculate the total for each bucket
    #np.unique will return counts if specified as the second argument
    total_count = np.unique(dev_bins, return_counts=True)[1]

    #This line ensures the total object passed in is modified for output
    total[:] = total_count[:]
    
    #calculate the number accurate for each bin
    #to do this, we can simply index the dev_bins with a binary array 
    #    indicating whether each position was correct or not to extract the accurate values
    #We generate this array using == using previous pred_class
    #We then return unique counts in the same way
    accurate_binary = pred_class == dev_labels
    dev_bins_correct = dev_bins[accurate_binary]
    correct_count = np.unique(dev_bins_correct, return_counts=True)[1]

    #This line ensures the total object passed in is modified for output
    correct[:] = correct_count[:]
    
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i] / total[i]
    print('p(pred) is %.13f to %.13f    total = %3d    accuracy = %.3f' % (0 if i==0 else buckets[i-1], buckets[i], total[i], accuracy))
```

This model appears to be weakly calibrated classifier. We obviously do not have perfect calibration (because we would for example expect bin 3 to have > 90% accuracy), but accuracy does improve as predicted probability improves. Naive Bayes tends to be overconfident, meaning the posterior probability tends to be higher than the accuracy, which we can see is true for all bins (except the first which spans a huge range). This is the signature of a weakly calibrated classifier. The only exception is between bin0 and bin1. This is likely due to the fact bin0 has very few examples and the fact that there are only 4 possible accuracies for a bin with 3 examples (0, 1/3, 2/3, or 1). This bin also spans a very large range of probabilities, so that particular bin is not the best indicator of calibration. As soon as we have 29+ examples starting in bin1, the trend is pretty clear.

We will design new features to see if we can produce a Bernoulli Naive Bayes model with better performance.  Accuracy of a model based on the original features and the accuracy of the model based on the new features will be shown.


```python
def P12():

### STUDENT START ###
    
    '''
    This function will try many novels features in order to improve the Bernoulli Naive Bayes Model
    We will first determine useful features, then perform hyperparameter optimization
    '''
    
    #Set a reasonable alpha and binarize threshold for both baseline and NB
    #Make sure they are consistent for both baseline, and the model with additional features
    NB_baseline = BernoulliNB(alpha = 0.001, binarize=0.1)
    NB_baseline.fit(mini_train_data, mini_train_labels)
    accuracy_baseline = NB_baseline.score(dev_data, dev_labels)
    print("Baseline performance: ", accuracy_baseline)
    NB = BernoulliNB(alpha = 0.001, binarize=0.1)

    def test_routine(new_train_data, new_dev_data, print_message = 'Dev set accuracy improvement: '):
        NB.fit(new_train_data, mini_train_labels)
        improvement = np.round(NB.score(new_dev_data, dev_labels) - accuracy_baseline, 5)
        if improvement > 0.0001:
            print('\033[1m', print_message, improvement)
        else:
            print('\033[0m', print_message, improvement)
        return improvement
    
    print("\n\nDEV SET IMPROVEMENT REPORTED BELOW\n")
    #collect expanded features for re-use
    #decorator to do everything to dev that is done to train
    feature_collection_train = {}
    feature_collection_dev = {}
        
    def combine_features_from_collection(train_data, dev_data, funcs):
        '''
        Collects features into a separate train and dev sets
        '''
        train = np.copy(train_data)
        dev = np.copy(dev_data)
        for func in funcs:
            train = np.hstack([train, feature_collection_train[func]])
            dev = np.hstack([dev, feature_collection_dev[func]])
        return (train, dev)
    
    def do_to_dev_too(func):
        '''
        Decorator that applies the same transformation to train and dev
        Also stores features in feature collection
        '''
        def wrapper(*args, **kwargs):
            train_data = kwargs['train_data']
            dev_data = kwargs['dev_data']
            
            new_train_data = func(*args, train_data)
            new_dev_data = func(*args, dev_data)
            
            feature_collection_train[func.__name__] = new_train_data
            feature_collection_dev[func.__name__] = new_dev_data

            return (np.append(train_data, new_train_data, axis=1), 
                    np.append(dev_data, new_dev_data, axis=1))
        return wrapper

    #Try adding a feature by summing across rows and columns
    #Won't do anything because always binarized to 1
    @do_to_dev_too
    def row_col_sums(data):
        colSums = np.sum(data.reshape([1000,28,28]), axis = 1)
        rowSums = np.sum(data.reshape([1000,28,28]), axis = 2)
        return np.hstack([colSums, rowSums])

    #Perform train/dev; will be used multiple times for each new feature
    train_data, deve_data = row_col_sums(train_data = mini_train_data, dev_data = dev_data)    
    test_routine(train_data, deve_data, "Row and Col sums:")

    #Try adding a feature by averaging across rows and columns
    @do_to_dev_too
    def row_col_means(data):
        colMeans = np.mean(data.reshape([1000,28,28]), axis = 1)
        rowMeans = np.mean(data.reshape([1000,28,28]), axis = 2)
        return np.hstack([colMeans, rowMeans])
        
    train_data, deve_data = row_col_means(train_data = mini_train_data, dev_data = dev_data)    
    test_routine(train_data, deve_data, "Row and Col means:")
    
    #Try adding just the row means
    @do_to_dev_too
    def row_means(data):
        rowMeans = np.mean(data.reshape([1000,28,28]), axis = 2)
        return rowMeans
        
    train_data, deve_data = row_means(train_data = mini_train_data, dev_data = dev_data)    
    test_routine(train_data, deve_data, "Row means:")
    
    #Try only using column maeans
    @do_to_dev_too
    def col_means(data):
        colMeans = np.mean(data.reshape([1000,28,28]), axis = 1)
        return colMeans
        
    train_data, deve_data = col_means(train_data = mini_train_data, dev_data = dev_data)    
    test_routine(train_data, deve_data, "Col means:")    
    
    #The idea is that some depending on the digit, the number of times a pixel is different
    #    from its neighboring pixel is different. For example, if a pixel is black and a neighbor is white
    #    this will be called a binary transition
    #Some digits like 1 and 7 will have fewer transitions than others, such as 8
    #This is a measure of "surface area" of the digit
    
    @do_to_dev_too
    def binary_transitions(data):
        binary_data = np.where(data > 0.1, 1, 0).reshape(1000, 28, 28)
        
        #This routine checks if all neighbors are the same
        #If so, then there's a transition. We will also take the mean
        #Since the features are binarized later, everything not 0 will be binarized to 1
        output = np.mean([binary_data[:, 1:27, 1:27] != binary_data[:, 1+i:27+i, 1+j:27+j] for i in [-1,0,1] for j in [-1,0,1]], axis=0)    
        return output.reshape(1000, -1)
    
    train_data, deve_data = binary_transitions(train_data = mini_train_data, dev_data = dev_data) 
    test_routine(train_data, deve_data, "Binary transitions:")

    #same binary transitions as above, except rather than using ONLY 0 and 1 to represent transitions,
    #    we will consider a transition as 1 only if it passes a threshold mean
    #We will also generate multiple features by setting the threshold at different values
    #The number of thresholds is a hyperparameter
    @do_to_dev_too
    def binary_transitions_threshold(threshold_nums, data):
        binary_data = np.where(data > 0.1, 1, 0).reshape(1000, 28, 28)
        
        output = np.mean([binary_data[:, 1:27, 1:27] != binary_data[:, 1+i:27+i, 1+j:27+j] for i in [-1,0,1] for j in [-1,0,1]], axis=0)    
        thresholds = np.sum(output > 0, axis=(1,2))
        thresholds.sort(axis=0)

        output_array = np.array([thresholds > threshold for threshold in thresholds[::threshold_nums]]).astype(np.int)
        return output_array.reshape(1000,-1)
    
    for threshold in [20,30,60]:
        train_data, deve_data = binary_transitions_threshold(threshold, train_data = mini_train_data, dev_data = dev_data) 
        test_routine(train_data, deve_data, "Binary transitions thresholded at %d:" %threshold)

    #Let's try the previous example with row means
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['row_means', 'binary_transitions_threshold']) 
    test_routine(train_data, deve_data, "Thresholded at 20, and row means:")
    
    #Try adding a feature by averaging across row blocks
    @do_to_dev_too
    def average_block_row_sizes(size, data):
        assert size < 28
        data_square = data.reshape([1000, 28, 28])
        max_iter_size = 28 - size + 1
        block_sums = np.empty(shape=[1000,0])

        i = 0
        j = 0
        while i < max_iter_size:
            block_sum = np.zeros([1000, 1])
            while j < max_iter_size:
                block_sum += np.sum(data_square[:, i:i+size, j:j+size], axis=(1,2)).reshape(1000,1)
                j+=1
            block_sums = np.hstack([block_sums, block_sum.reshape(1000,1)])
            i+=1
        return block_sums.reshape(1000, -1) / max_iter_size

    for block_size in [2,6]:
        train_data, deve_data = average_block_row_sizes(block_size, train_data = mini_train_data, dev_data = dev_data) 
        test_routine(train_data, deve_data, "Row averaging blocks of %d:" %block_size)
    
    #Try adding a feature by averaging across column blocks
    @do_to_dev_too
    def average_block_column_sizes(size, data):
        assert size < 28
        data_square = data.reshape([1000, 28, 28])
               
        max_iter_size = 28 - size + 1
        block_sums = np.empty(shape=[1000,0])

        i = 0
        j = 0
        while i < max_iter_size:
            block_sum = np.zeros([1000, 1])
            while j < max_iter_size:
                block_sum += np.sum(data_square[:, j:j+size, i:i+size], axis=(1,2)).reshape(1000,1)
                j+=1
            block_sums = np.hstack([block_sums, block_sum.reshape(1000,1)])
            i+=1
        return block_sums.reshape(1000, -1) / max_iter_size

    for block_size in [2,10]:
        train_data, deve_data = average_block_column_sizes(block_size, train_data = mini_train_data, dev_data = dev_data) 
        test_routine(train_data, deve_data, "Column averaging blocks of %d:" %block_size)

    #We will try enclosure
    @do_to_dev_too
    def enclosure(data):
        binary_data = np.where(data > 0.1, 1, 0).reshape(1000, 28, 28)
        
        output = np.zeros(binary_data.shape)
        
        for n, image in enumerate(binary_data):
            for row in range(0,28):
                for col in range(0,28):
                    if image[row, col] == 1:
                        output[n][row][col] = 0
                    else:
                        #Generate if a transition occurs above, below, left, and right of pixel
                        #If so, set int as 1
                        upper = np.sum(image[:row, col])
                        lower = np.sum(image[row:, col])
                        left = np.sum(image[row, :col])
                        right = np.sum(image[row, col:])
                        output[n][row][col] = np.all([upper, lower, left, right]).astype(np.int)
        return np.reshape(output, [1000,-1])
    
    train_data, deve_data = enclosure(train_data = mini_train_data, dev_data = dev_data) 
    test_routine(train_data, deve_data, "Enclosure:")
    
    #Let's try enclosure with thresholding
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['enclosure','binary_transitions_threshold']) 
    test_routine(train_data, deve_data, "Thresholded at 20 and enclosure:")
    
    #We will try to measure length of contiguous stretches of 1 pixels
    #We will measure in blocks of 2, meaning if we are checking across a row,
    #    then we will treat the first 2 pixels as a single row
    #This allows some slant in the writing.
    #The block size is a hyperparameter
    @do_to_dev_too
    def contiguous_row_stretches(block_size, data):
        binary_data = np.where(data > 0.1, 1, 0).reshape(1000, 28, 28)
        
        contiguous_stretches = np.zeros([1000,28//block_size])
        
        for n, image in enumerate(binary_data):
            for row in range(0,28,block_size):
                
                max_stretch = 0
                stretch = 0
                for i in image[row:row+block_size,:].max(axis=0):
                    if i == 0:
                        max_stretch = max(stretch, max_stretch)
                        stretch = 0
                    else:
                        stretch += 1
                        
                max_stretch = max(stretch, max_stretch)
                contiguous_stretches[n, row//block_size] = max_stretch
        
        output = np.empty((1000,0))
        for threshold in np.unique(contiguous_stretches):
            output = np.hstack([output, contiguous_stretches > threshold])
            
        return np.reshape(output, (1000, -1))
    
    train_data, deve_data = contiguous_row_stretches(2, train_data = mini_train_data, dev_data = dev_data) 
    test_routine(train_data, deve_data, "Contiguous row stretches:")

    #Do the same thing, except for columns
    @do_to_dev_too
    def contiguous_col_stretches(block_size, data):
        binary_data = np.where(data > 0.1, 1, 0).reshape(1000, 28, 28)
        
        contiguous_stretches = np.zeros([1000,28//block_size])
        
        for n, image in enumerate(binary_data):
            for col in range(0,28,block_size):
                
                max_stretch = 0
                stretch = 0
                for i in image[:,col:col+block_size].max(axis=1):
                    if i == 0:
                        max_stretch = max(stretch, max_stretch)
                        stretch = 0
                    else:
                        stretch += 1
                        
                max_stretch = max(stretch, max_stretch)
                contiguous_stretches[n, col//block_size] = max_stretch
        
        output = np.empty((1000,0))
        for threshold in np.unique(contiguous_stretches):
            output = np.hstack([output, contiguous_stretches > threshold])
            
        return np.reshape(output, (1000, -1))
    
    train_data, deve_data = contiguous_col_stretches(2, train_data = mini_train_data, dev_data = dev_data) 
    test_routine(train_data, deve_data, "Contiguous col stretches:")
    
    #First, we'll see if using entropy filtering helps
    #The procedure we will take is as follows
    #Train models on only a single feature at a time
    #Then calculate the entropy using the posterior distribution: P(Y_j | x_i)
    #Eliminate features below a certain threshold
    
    #########Entropy filtering###########
    def entropy(probs):
        probs = probs [probs > 0]
        return -np.sum(np.log2(probs) * probs)
    
    num_features = 784 
    single_example = np.array([1]).reshape(1, -1)
    entropies = np.zeros(num_features)
    probs = []
    threshold = 2
    for feat_index in range(num_features):

        data_with_only_single_feature = mini_train_data[:, [feat_index]]
        if np.sum(data_with_only_single_feature) == 0:
            #entropies[feat_index] = threshold - 1
            continue

        NB.fit(data_with_only_single_feature, mini_train_labels)
        probs_class = NB.predict_proba(single_example)

        entropies[feat_index] = entropy(probs_class)

    mini_train_data_entropy = mini_train_data[:, entropies > threshold]
    dev_data_entropy = dev_data[:, entropies > threshold]
    test_routine(mini_train_data_entropy, dev_data_entropy, "Entropy filter:")
    ##########Entropy filtering############
    
    #Let's try contiguous row and column blocks
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['contiguous_row_stretches','contiguous_col_stretches']) 
    test_routine(train_data, deve_data, "Contiguous row and col stretches:")
    
    #Let's try contiguous row and column blocks
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['enclosure','binary_transitions_threshold','contiguous_row_stretches']) 
    test_routine(train_data, deve_data, "Three sets:")
    
    #Let's try contiguous row and column blocks
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['enclosure','binary_transitions_threshold','contiguous_row_stretches','contiguous_col_stretches']) 
    test_routine(train_data, deve_data, "Four sets:")
    #It seems like the last combination of features is the best so far, with a 4.7% increase
    #We will perform some hyperparameter optimization for the three different addition set of features
    #Enclosure does not change, but binary_transitions_threshold, contiguous_row_stretches, contiguous_col_stretches all have hyperparameters
    
    NB.fit(train_data, mini_train_labels)
    print("\nBest performance so far:", NB.score(deve_data, dev_labels))
    
    print('\033[0m'+"\nHYPERPARAMETER OPTIMIZATION\n")
    print('binary_transitions_threshold','contiguous_row_stretches','contiguous_col_stretches','dev_improvement')
    bt_hparams = np.linspace(15,150,150/15).astype(int)
    crs_hparams = [2, 4, 7]
    ccs_hparams = [2, 4, 7]
    
    best_params = [0,0,0]
    best_performance = 0
    for bt in bt_hparams:
        for crs in crs_hparams:
            for ccs in ccs_hparams:
                train_data, deve_data = binary_transitions_threshold(bt, train_data = mini_train_data, dev_data = dev_data)
                train_data, deve_data = contiguous_row_stretches(crs, train_data = mini_train_data, dev_data = dev_data) 
                train_data, deve_data = contiguous_col_stretches(ccs, train_data = mini_train_data, dev_data = dev_data) 
                train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['enclosure','binary_transitions_threshold','contiguous_row_stretches','contiguous_col_stretches'])
                dev_set_performance = test_routine(train_data, deve_data, ' '.join(list(map(str, [bt, crs, ccs]))))
                if dev_set_performance > best_performance:
                    best_performance = dev_set_performance
                    best_params = [bt, crs, ccs]
    print('\033[0m',"Best params:", best_params)
    
    print('\033[1m','\n\nFINAL MODEL','\033[0m')
    train_data, deve_data = binary_transitions_threshold(best_params[0], train_data = mini_train_data, dev_data = dev_data)
    train_data, deve_data = contiguous_row_stretches(best_params[1], train_data = mini_train_data, dev_data = dev_data) 
    train_data, deve_data = contiguous_col_stretches(best_params[2], train_data = mini_train_data, dev_data = dev_data) 
    train_data, deve_data = combine_features_from_collection(mini_train_data, dev_data, ['enclosure','binary_transitions_threshold','contiguous_row_stretches','contiguous_col_stretches']) 
    NB.fit(train_data, mini_train_labels)
    print('\033[1m', "Final performance:", NB.score(deve_data, dev_labels))
    test_routine(train_data, deve_data, 'Final model improvement on dev set:')
### STUDENT END ###
P12()
```
